```python
input_list = ['x','y','z']
result = [ item*num for item in input_list for num in range(1,5)  ]
print("['x','y','z'] => " +   str(result))
```

```python
input_list = ['x','y','z']
result = [ item*num for item in range(1,5) for num in input_list]
print("['x','y','z'] => " +   str(result))
```

```python
input_list = [2,3,4]
input_list1 = [2,3,4,5]
result = [ [item+num] for item in input_list for num in range(0,3)]
result1 = [ [item+num for item in input_list1] for num in range(0,2)]
print("[2,3,4] =>" +  str(result) + ',' + str(result1))
```
